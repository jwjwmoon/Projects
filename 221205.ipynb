### z-score method 이상치제거

```python
def not_outlier(data, col, z):
    return data[abs(data[col] - np.mean(data[col]))/np.std(data[col]) <= z].index
    
```

```python
new_data = data.loc[not_outlier(data, '무역수지', 1.5)]
new_data
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].hist(data)
axs[0].set_title('Original Data')

axs[1].hist(new_data)
axs[1].set_title('New Data')

plt.show()
```

### 정규화(normalization) vs. 표준화(standardization)

1. 스케일링 시 최댓값, 최솟값이 사용됨   /  스케일링 시 표준편차가 사용됨
2. 피처의 크기가 다를 때 사용          / 평균0, 표준편차가 1이 되도록
3. [0,1]or[-1,1] 사이 값으로 스케일링 / 특정 범위로 제한되지 않음
4. 분포에 대해 모를 때 유용           / 피처가 정규분포인 경우 유용
5. MinMaxScaler, Normalizer     / StandardScaler, RobustScaler

### 범주형 데이터 vs. 수치형 데이터

1. 범주형 데이터 : 몇 개의 범주로 나누어진 자료  ex) 성별, 혈액형 등
2. 수치형 데이터 : 크기나 가치 비교 가능
- (1) 이산형 자료 : 불연속적                ex) 일정기간 동안의 발생횟수
- (2) 연속형 자료 : 연속적인 값을 갖는 데이터   ex) 신장, 체중 혈압

#### Q) Should we apply normalization to test data as well? 

A) Yes.

That is because your model works on the representation given by its input vectors. The scale of those numbers is 
part of the representation. This is a bit like converting between feet and metres ... 
a model or formula would work with just one type of unit normally.

Not only do you need normalization, but you should apply the exact same scaling as for your training data. 
That means storing the scale and offset used with your training data, and using that again. 
A common beginner mistake is to separately normalize your train and test data.
